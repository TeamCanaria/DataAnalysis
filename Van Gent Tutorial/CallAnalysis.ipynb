```python
#to call the heart rate analysis module:
import PVGHeartRateAnalysis as hra

dataset = hra.get_data('2018-01-25_Finger.csv')
hra.process(dataset, 0.75, 2720)

#We have imported the above Python module as an object called 'hra'
#this object contains the dictorionary 'measures' which contains all our values
#now we can also retrieve the BPM value, and later other values, like so:
bpm = hra.measures['bpm']
ibi = hra.measures['ibi']
sdnn = hra.measures['sdnn']

#to view all objects in the dictionary, use "keys()" like so:
print(hra.measures.keys())
```
