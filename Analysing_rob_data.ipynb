## Preprocess the data

```python
rob_data = pd.read_csv('Rob_data.csv', delimiter=',')
```

```python
rob_data.head(20)
```

```python
# Set the columns ts as the index of dataframe, 
# this make the data selection more convenience
rob_data['ts'] = pd.to_datetime(rob_data['ts'])
rob_data.index = rob_data['ts']
del rob_data['ts']
```

```python
rob_data.tail(10)
```

```python
rob_data.head(10)
```

The signal pattern which we are interested is fatigued, The fatigured pattern will be used to compare with sleep and awake signal to find the different. The time "Wed 1 Nov 2017 3:30" is chosen as the reference signal pattern for fatigured as at this time, Rob was clearly exhausted.

```python
# Convert normal datetime into the format of pandas
print(pd.to_datetime("Fri 20 Oct 2017"))    # Data choose to check the fatigued data
print(pd.to_datetime("Sat 14 Oct 2017"))        # Data choose to check the awaking data
```

```python
rob_data['2017']
```

```python
# The data from "Fri 13 Oct 2017" missing value for early morning to 1:27 pm until midnight,
# we only need data at night from 22:43 as this is Rob's bedtime
# Count the number of rows which have missing values
counter = 0
sleep_data = rob_data['2017'] 
for index, row in sleep_data.iterrows():
    if (row.ir == 0) and (row.red == 0):
        counter = counter + 1
    else:
        break
counter
```

```python
# Remove all Missing data
sleep_data = sleep_data[233:]
sleep_data.head(20)
```

```python
# The data from Sat 14 Oct 2017 is missing data at night,
# we only need data in the morning from 6:43 am as this time Rob was awake
# Count number of missing value 
counter = 0
awake_data = rob_data['2017-10-14'] 
for index, row in awake_data[::-1].iterrows():
    if (row.ir == 0) and (row.red == 0):
        counter = counter + 1
    else:
        break
counter
```

```python
# Remove all missing values
awake_data = awake_data[:-6000]
awake_data.head(20)
```

```python
# Find the length of two dataset
print('sleep data length: ' + str((sleep_data.iloc[[-1]].index - sleep_data.iloc[[0]].index)[0]))
print('awake data length: ' + str((awake_data.iloc[[-1]].index - awake_data.iloc[[0]].index)[0]))
```

The length of sleep dataset is 22 hours 32 minutes, the length of awake dataset is 15 hours 24 minutes
